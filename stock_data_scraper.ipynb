Fetching stock data of biggest players in the market

```python
def fetch_multiple_stocks(api_key, symbols):
    """
   
    Parameters:
    api_key (str): Alpha Vantage API key
    symbols (list): List of stock symbols
    
    Returns:
    dict: Dictionary of DataFrames with stock symbols as keys
    """
    symbols =['AMZN', 'XOM', 'GOOGL', 'MSFT', 'LIN', 'WMT', 'NEE', 'LLY', 'BRK.B', 'PLD', 'GE']

    base_url = 'https://www.alphavantage.co/query'
    stock_data = {}
    
    for symbol in symbols:
        params = {
            'function': 'TIME_SERIES_DAILY',
            'symbol': symbol,
            'outputsize': 'full',
            'datatype': 'csv',
            'apikey': 'KARS1PN9N573AELO'
        }
        
        try:
            response = requests.get(base_url, params=params)
            if response.status_code == 200:
                df = pd.read_csv(io.StringIO(response.text))
                stock_data[symbol] = df
                print(f"Successfully fetched data for {symbol}")
                # Alpha Vantage has a rate limit of 5 API calls per minute for free tier
                time.sleep(12)  # Adding delay to respect rate limits
            else:
                print(f"Failed to fetch data for {symbol}: Status code {response.status_code}")
                
        except Exception as e:
            print(f"Error fetching data for {symbol}: {str(e)}")
    
    return stock_data

# Usage example
api_key = 'KARS1PN9N573AELO'  # Replace with your API key
symbols = ['AMZN', 'XOM', 'GOOGL', 'MSFT', 'LIN', 
           'WMT', 'NEE', 'LLY', 'BRK.B', 'PLD', 'GE']

# Import required libraries if not already imported
import requests
import pandas as pd
import io
import time

# Fetch the data
stock_datasets = fetch_multiple_stocks(api_key, symbols)
```

```python
# Add symbol column to each DataFrame
for symbol, df in stock_datasets.items():
    df['symbol'] = symbol  # Add symbol column

stock_datasets
```

```python
# Concatenate all DataFrames into a single DataFrame

combined_df = pd.concat(stock_datasets.values(), ignore_index=True)

# Sort by symbol and timestamp
combined_df = combined_df.sort_values(['symbol', 'timestamp'])
```

```python
combined_df['timestamp'] = pd.to_datetime(combined_df['timestamp'])
combined_df
```

```python
combined_df= combined_df.reset_index(drop=True)
```

```python
combined_df = combined_df.drop('index', axis=1)
combined_df
```

```python
# Save to CSV
combined_df.to_csv('stocks_historical_data.csv', index=False)
```
